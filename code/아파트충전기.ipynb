```python
import pandas as pd
import os
import numpy as np
```

```python
DF2_FILENAME = '20220925_아파트정보목록_위경도 추가.csv'
DF2_FILEPATH = os.path.join(os.path.dirname(os.getcwd()),'data','dataset',DF2_FILENAME)
df = pd.read_csv(DF2_FILEPATH)
df
```

```python
df1 = df.drop(['Unnamed: 0','index'],axis=1)
df1
```

```python
df1['자동차수']= df1['세대수'] * 1.1
df1
```

```python
apt = df1[df1['자동차수']>=100 ]
apt.reset_index(inplace=True)
apt
```

```python
apt_1=apt.copy()
apt_1
```

```python
date = apt_1['사용검사일']
date
```

```python
apt_1['사용검사일'] = apt_1['사용검사일'].apply(lambda x:''.join(x.split('.')))
apt_1
```

```python
apt_1['사용검사일'] = apt_1['사용검사일'].apply(lambda x:''.join(x.split(' ')))
apt_1
```

```python
import re
regex = "\(.*\)|\s-\s.*"
for i in range(len(apt_1)):
    apt_1['사용검사일'][i] = re.sub(regex, '', apt_1['사용검사일'][i])
```

```python
apt_1['사용검사일'] = pd.to_numeric(apt_1['사용검사일'])
apt_1
```

```python
apt_1.loc[1, '사용검사일'] = 20230730
apt_1
```

```python
#구축
apt_2 = apt_1[apt_1['사용검사일']<= 20220128]
apt_2
```

```python
#신축
apt_3 = apt_1[apt_1['사용검사일']> 20220128]
apt_3
```

```python
#충전기수 : 올림
apt_2['충전기수']= apt_2['자동차수'] * 0.02
apt_2['충전기수']= apt_2['충전기수'].apply(np.ceil)
```

```python
apt_3['충전기수']= apt_3['자동차수'] * 0.02
apt_3['충전기수']= apt_3['충전기수'].apply(np.ceil)
```

```python
fin_apt = pd.concat([apt_2,apt_3])
fin_apt.reset_index(inplace=True)
```

```python
fin_apt
```

```python
fin_apt = fin_apt.drop(['level_0','index'],axis=1)
fin_apt
```

```python
fin_apt.to_csv( '../data/dataset/20220925_아파트정보목록_위경도_충전기수 추가.csv')
```
