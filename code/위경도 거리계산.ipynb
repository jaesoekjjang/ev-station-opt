# 기존 충전소 위경도

```python
import os
import pandas as pd

DF1_FILENAME = '20220925_아파트정보목록_위경도_충전기수 추가.csv'
DF1_FILEPATH = os.path.join(os.path.dirname(os.getcwd()),'data','dataset',DF1_FILENAME)

DF2_FILENAME = 'public_parking_lot_50대이상.csv'
DF2_FILEPATH = os.path.join(os.path.dirname(os.getcwd()),'data','dataset',DF2_FILENAME)

DF3_FILENAME = '전기충전소정보.xml'
DF3_FILEPATH = os.path.join(os.path.dirname(os.getcwd()),'data','dataset',DF3_FILENAME)

df1 = pd.read_csv(DF1_FILEPATH)
df2 = pd.read_csv(DF2_FILEPATH)
station_df = pd.read_xml(DF3_FILEPATH)
```

```python
#용인시 충전기
addr_series = station_df['addr'].str.contains('용인시', na=False)
addr_series.isna().sum()
df3 = station_df[addr_series]
```

```python
#아파트 충전기 정보
df1
```

```python
#공영주차장
df2
```

```python
#기존 충전소 정보
df3
```

```python
df3.rename(columns={'lat':'위도','lng':'경도'},inplace=True)
df3
```

```python
df1.rename(columns={'단지명':'이름'},inplace=True)
df2.rename(columns={'주차장명':'이름'},inplace=True)
df3.rename(columns={'statNm':'이름'},inplace=True)
```

```python
df1 = df1[['이름','위도','경도']]
df2 = df2[['이름','위도','경도']]
df3 = df3[['이름','위도','경도']]
```

```python
df1 = df1.reset_index()
df2 = df2.reset_index()
df3 = df3.reset_index()
```

```python
cha = pd.concat([df1,df2,df3],axis=0)
cha.reset_index(inplace=True)
cha
```

```python
fin_cha = cha.drop(['level_0','index'],axis=1)
fin_cha.drop_duplicates(['위도', '경도'])
```

```python
fin_cha.to_csv( '../data/dataset/기존충전소위경도정보.csv')
```
